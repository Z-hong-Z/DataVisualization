```python
from pyecharts import Pie
import csv
datax=[]
datay=[]
filename = "vote_result.csv"
with open(filename,encoding='utf-8') as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datax.append(datarow[0])
            datay.append(datarow[1])
pie = Pie("饼图示例")
pie.add("", datax,datay, is_label_show=True)
pie.render()
```

```python
#去掉图例
pie.add("", datax,datay, is_label_show=True,is_legend_show=False)
pie.render()
```

```python
#图例移到左边，标题正中显示
pie = Pie("饼图示例",title_pos='center')
pie.add("", datax,datay,legend_orient="vertical",legend_pos="left", is_label_show=True)
pie.render()
```

```python
#不显示标签
pie = Pie("饼图示例")
pie.add("", datax,datay, is_label_show=False)
pie.render()
```

```python
#添加标题和副标题
pie = Pie("数据可视化-用户感兴趣领域",
         "以下是读者的投票结果。\n"
         "读者对金融，医疗保健，市场业3个领域最感兴趣。",title_pos='left')
pie.add("", datax,datay,legend_orient="vertical",legend_pos="right", is_label_show=True)
pie.render()
```

```python
from pyecharts import Pie
import csv
datax=[]
datay=[]
filename = "vote_result.csv"
with open(filename,encoding='utf-8') as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datax.append(datarow[0])
            datay.append(datarow[1])
pie = Pie("饼图示例")
pie.add("", datax,datay,
       center=[60,60],legend_orient="vertical",radius=[30,75],legend_pos='right',is_label_show=True)
pie.render()
```

```python
from pyecharts import Bar
import csv
data1=[]
data2=[]
data3=[]
data4=[]
filename = "presidential_approval_rate.csv"
with open(filename,encoding='utf-8') as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            data1.append(datarow[0])
            data2.append(datarow[1])
            data3.append(datarow[2])
            data4.append(datarow[3])
bar = Bar("柱状图数据堆叠示例",width="10px")
bar.add("支持",data1,data2,is_stack=True,xaxis_rotate=70)
bar.add("反对",data1,data3,is_stack=True,xaxis_rotate=70)
bar.add("不发表意见",data1,data4,is_stack=True,xaxis_rotate=70)
bar.render()
```

```python
from pyecharts import Bar
import pandas as pd
pre_approval_rate = pd.read_csv('presidential_approval_rate.csv')
bar = Bar("柱状图数据堆叠示例",title_pos='center')
list_support = ['支持','反对','不发表意见']
for i in range(pre_approval_rate.iloc[:,0].size):
    issue = pre_approval_rate.loc[i,'political_issue']
    bar.add(issue, list_support, pre_approval_rate.loc[i,['support','oppose','no_opinion']], legend_orient="vertical",legend_pos="right",is_stack=False)
bar.render()
```

```python
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import squarify
color = sns.color_palette()
pd.options.mode.chained_assignment = None  # default='warn'
products_df = pd.read_csv('products.csv')
aisles_df = pd.read_csv('aisles.csv')
departments_df = pd.read_csv('departments.csv')
order_products_prior_df = pd.merge(products_df, aisles_df, on='aisle_id', how='left')
order_products_prior_df = pd.merge(order_products_prior_df, departments_df, on='department_id', how='left')
order_products_prior_df.head()
temp = order_products_prior_df[['product_name', 'aisle', 'department']]
temp = pd.concat([
    order_products_prior_df.groupby('department')['product_name'].nunique().rename('products_department'),
    order_products_prior_df.groupby('department')['aisle'].nunique().rename('aisle_department')
], axis=1).reset_index()
temp = temp.set_index('department')
temp2 = temp.sort_values(by="aisle_department", ascending=False)
print(temp)
print(temp2)
x = 0
y = 0
width = 100
height = 100
cmap = matplotlib.cm.viridis
mini, maxi = temp2.products_department.min(), temp2.products_department.max()
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in temp2.products_department]
colors[1] = "#FBFCFE"
labels = ["%s\n%d aisle num\n%d products num" % (label) for label in
          zip(temp2.index, temp2.aisle_department, temp2.products_department)]
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, aspect="equal")
ax = squarify.plot(temp2.aisle_department, color=colors, label=labels, ax=ax, alpha=.7)
fig.suptitle("How are aisles organized within departments", fontsize=20 )
ax.set_xticks([])
ax.set_yticks([])
img = plt.imshow([temp2.products_department], cmap=cmap)
img.set_visible(False)
fig.colorbar(img, orientation="vertical", shrink=.96)
fig.text(.76, .9, "numbers of products", fontsize=14)
plt.show()
```
